```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris['Species'].value_counts()
```

```python
iris.drop(['Id'],axis =1,inplace = True)
```

balanced dataset by looking at the class lable it is equally distributed if not it is imbalance dataset

```python
iris.shape
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='PetalLengthCm',y = 'PetalWidthCm',data = iris,hue = 'Species')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x='SepalLengthCm',y = 'SepalWidthCm',data = iris,hue = 'Species')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.pairplot(data = iris,hue = 'Species',height =2.5)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.histplot(iris,x='SepalLengthCm',hue='Species')
```

```python
sns.set_style("whitegrid")
sns.histplot(iris,x='SepalLengthCm',hue='Species',kde=True)
```

```python
sns.set_style("darkgrid")
sns.displot(iris,x='PetalLengthCm',hue='Species',kde = True)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(iris,x='SepalWidthCm',hue='Species',kde = True)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.boxplot(data = iris,x='Species',y="PetalLengthCm",width=0.6,linewidth=2)
plt.show()
```

violinplot

```python
sns.violinplot(data=iris,x='Species',y="PetalLengthCm",width=0.9,linewidth=2.5)
plt.show()
```
